### Matplotlib
- simple to advanced plotting capabilities for quick, easy visualization
- 2D and 3D capabilities
- reproducible, automated figure generation
- https://matplotlib.org/gallery/index.html



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.0, 2, 0.01) # range from 0 to 2 at intervals of 0.01
y1 = np.sin(2 * np.pi * x)  # sin(2xpi)
y2 = 1.2 * np.sin(4 * np.pi * x) # 1.2 sin(4xpi)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(12,6))

# Plot 1
ax1.fill_between(x, 0, y1)
ax1.set_ylabel('between y1 and 0')

# Plot 2
ax2.fill_between(x, y1, 1)
ax2.set_ylabel('between y1 and 1')

# Plot 3
ax3.fill_between(x, y1, y2)
ax3.set_ylabel('between y1 and y2')
ax3.set_xlabel('x')
```
